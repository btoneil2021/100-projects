```python
import matplotlib.pyplot as plt
```

```python
def create_bar_chart(x_data, y_data, title, x_label, y_label):
    """
    Generates and displays a simple bar chart.

    Args:
        x_data (list): A list of categories for the x-axis.
        y_data (list): A list of numerical values for the y-axis.
        title (str): The title of the chart.
        x_label (str): The label for the x-axis.
        y_label (str): The label for the y-axis.
    """
    plt.figure()
    plt.bar(x_data, y_data)
    plt.title(title)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.show()
```

```python
def create_line_chart(x_data, y_data, title, x_label, y_label):
    """
    Generates and displays a simple line chart.

    Args:
        x_data (list): A list of numerical values for the x-axis.
        y_data (list): A list of numerical values for the y-axis.
        title (str): The title of the chart.
        x_label (str): The label for the x-axis.
        y_label (str): The label for the y-axis.
    """
    plt.figure()
    plt.plot(x_data, y_data)
    plt.title(title)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.grid(True)
    plt.show()
```

```python
def create_scatter_plot(x_data, y_data, title, x_label, y_label):
    """
    Generates and displays a simple scatter plot.

    Args:
        x_data (list): A list of numerical values for the x-axis.
        y_data (list): A list of numerical values for the y-axis.
        title (str): The title of the chart.
        x_label (str): The label for the x-axis.
        y_label (str): The label for the y-axis.
    """
    plt.figure()
    plt.scatter(x_data, y_data)
    plt.title(title)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.grid(True)
    plt.show()
```

```python
# --- 1. Bar Chart Example ---
product_categories = ['A', 'B', 'C', 'D']
sales_numbers = [150, 230, 95, 410]
create_bar_chart(
    x_data=product_categories,
    y_data=sales_numbers,
    title='Sales by Product Category',
    x_label='Product Category',
    y_label='Total Sales'
)
```

```python
# --- 2. Line Chart Example ---
months = [1, 2, 3, 4, 5, 6]
stock_prices = [100, 105, 115, 110, 125, 132]
create_line_chart(
    x_data=months,
    y_data=stock_prices,
    title='Stock Price Over Time',
    x_label='Month',
    y_label='Price (USD)'
)
```

```python
# --- 3. Scatter Plot Example ---
study_hours = [2, 3, 5, 1, 6, 4, 7]
exam_scores = [65, 70, 85, 50, 90, 75, 95]
create_scatter_plot(
    x_data=study_hours,
    y_data=exam_scores,
    title='Exam Score vs. Hours Studied',
    x_label='Hours Studied',
    y_label='Exam Score (%)'
)
```
