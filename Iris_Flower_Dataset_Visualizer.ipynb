```python
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
```

```python
iris = load_iris()
X = iris.data
y = iris.target
target_names = iris.target_names
```

```python
print(X.shape)
print(y.shape)
```

```python
x_feature_index = 0
y_feature_index = 1

# 4. Create the scatter plot
# Create a figure and an axes object
fig, ax = plt.subplots()

# Define colors for each species
colors = ['navy', 'turquoise', 'darkorange']

# Iterate through each species (0, 1, 2) and plot their data points
for i, color, target_name in zip(range(len(target_names)), colors, target_names):
    # Select the data points for the current species
    # X[y == i, x_feature_index] gives the sepal length for all flowers of species i
    # X[y == i, y_feature_index] gives the sepal width for all flowers of species i
    ax.scatter(X[y == i, x_feature_index], X[y == i, y_feature_index], color=color, alpha=0.8, label=target_name)

# 5. Customize the plot
# Set the title of the plot
ax.set_title('Iris Dataset: Sepal Length vs. Sepal Width')
# Set the label for the x-axis
ax.set_xlabel(iris.feature_names[x_feature_index])
# Set the label for the y-axis
ax.set_ylabel(iris.feature_names[y_feature_index])
# Add a legend to the plot to identify the species
ax.legend()
# Add a grid for better readability
ax.grid(True)

# 6. Display the plot
plt.show()
```
