```python
pathToBeSorted = r'/content/data-analysis-template-main'
destDir = r'/content/data-analysis-template-main-sorted'
```

```python
if not os.path.exists(destDir):
  os.makedirs(destDir)
```

```python
def moveFile(item_path_arg, destDir_arg=destDir):
  # Os.path.basename - grabs just the name of the file (or last directory in path)
  item_name = os.path.basename(item_path_arg)

  print(f"Processing File: {item_name}")

  # os.path.splitext - splits into the filename and an extension (with a dot)
  filename_without_path, file_extension_w_dot = os.path.splitext(item_name)
  file_extension = ""

  if file_extension_w_dot:
    file_extension = file_extension_w_dot[1:].lower()

  if not file_extension:
    print(f"Skipping file with unknown extension: {item_name}")
    return

  print(f"File: {item_name} - Extension: {file_extension}")

  dest_dir_path = os.path.join(destDir_arg, file_extension)

  try:
    # os.makedirs creates all necessary directories along a specific path
    # Helpful here because we are going deep into different levels of directories
    os.makedirs(dest_dir_path, exist_ok=True)
    print(f"Moving file from {item_path_arg} to: {dest_dir_path}")
    # shutil.move - move from a source path to a destination path
    shutil.move(item_path_arg, dest_dir_path)
    print(f"Successfully moved '{item_name}")
  except FileNotFoundError:
    print(f"The source path {item_path_arg} was not found")
  except shutil.Error as e:
    print(f"Error moving file: {item_name} - {e}")
  except OSError as e:
    print(f"Error creating directory or moving file: {dest_dir_path} - {e}")
```

```python
def enterDir(currentBasePath, dir_to_scan_name=None):
  print(f"Current base path: {currentBasePath}, Directory to scan name: {dir_to_scan_name}")

  if dir_to_scan_name:
    path_to_scan = os.path.join(currentBasePath, dir_to_scan_name)
  else:
    path_to_scan = currentBasePath

  print(f"Starting to organize files in: {path_to_scan}")
  try:
    # os.listdir - lists everything at the given path
    all_items = os.listdir(path_to_scan)
    print(f"Found items: {all_items}")
  except FileNotFoundError:
    print(f"The path {path_to_scan} was not found")
    return
  except Exception as e:
    print(f"An error occured: {e}")
    return

  if not all_items:
    print("The path is empty")
    return

  for item_name in all_items:
    item_full_path = os.path.join(path_to_scan, item_name)
    # os.path.isfile - checks if the item at the path is a file
    if os.path.isfile(item_full_path):
      moveFile(item_full_path)
    # os.path.isdir - checks if the item at the path is a directory
    elif os.path.isdir(item_full_path):
      print(f"Entering subdirectory: {item_name} (Full path: {item_full_path})")
      enterDir(path_to_scan, item_name)
    else:
      print(f"Skipping Other: {item_name}")
    print()
```

```python
print(f"Starting script. Source: '{pathToBeSorted}', Destination: '{destDir}'")
enterDir(pathToBeSorted)
print("Script finished.")
```
